```python
import datetime

# Definir variables y listas
pisos = 10
departamentos_por_piso = 4
tipos_departamento = ['A', 'B', 'C', 'D']
precios = {'A': 3800, 'B': 3000, 'C': 2800, 'D': 3500}
departamentos_disponibles = [['{}{}'.format(tipo, piso) for tipo in tipos_departamento] for piso in range(1, pisos + 1)]
departamentos_vendidos = []
compradores = {}

# Función para mostrar el menú
def mostrar_menu():
    print("Menú:")
    print("1. Comprar departamento")
    print("2. Mostrar departamentos disponibles")
    print("3. Ver listado de compradores")
    print("4. Mostrar ganancias totales")
    print("5. Salir")

# Función para comprar departamento
def comprar_departamento():
    print("\nDepartamentos disponibles:")
    for piso, disponibles in enumerate(departamentos_disponibles, start=1):
        print("Piso {}: {}".format(piso, disponibles))

    departamento_seleccionado = input("Ingrese el departamento que desea comprar (ejemplo: A3): ").upper()

    if departamento_seleccionado in departamentos_vendidos:
        print("Departamento no disponible. Seleccione otro.")
        return

    tipo_departamento = departamento_seleccionado[0]
    piso_departamento = int(departamento_seleccionado[1:])

    if tipo_departamento not in tipos_departamento or piso_departamento < 1 or piso_departamento > pisos:
        print("Selección no válida. Intente nuevamente.")
        return

    run_comprador = input("Ingrese el RUN del comprador (sin guion ni puntos): ")
    compradores[run_comprador] = departamento_seleccionado
    departamentos_vendidos.append(departamento_seleccionado)
    print("Operación realizada correctamente.")

# Función para mostrar departamentos disponibles
def mostrar_departamentos_disponibles():
    print("\nDepartamentos disponibles:")
    for piso, disponibles in enumerate(departamentos_disponibles, start=1):
        print("Piso {}: {}".format(piso, disponibles))

# Función para ver listado de compradores
def ver_listado_compradores():
    print("\nListado de compradores:")
    for run, departamento in sorted(compradores.items(), key=lambda x: x[0]):
        print("RUN: {}, Departamento: {}".format(run, departamento))

# Función para mostrar ganancias totales
def mostrar_ganancias_totales():
    print("\nGanancias totales:")
    total_por_tipo = {tipo: 0 for tipo in tipos_departamento}

    for departamento in departamentos_vendidos:
        tipo_departamento = departamento[0]
        total_por_tipo[tipo_departamento] += precios[tipo_departamento]

    total_general = sum(total_por_tipo.values())

    for tipo, total_tipo in total_por_tipo.items():
        print("Tipo {}: {} UF {}".format(tipo, precios[tipo], total_tipo))

    print("TOTAL {} UF".format(total_general))

# Función para salir
def salir():
    fecha_actual = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print("\nSaliendo del sistema. Fecha y hora actual: {}".format(fecha_actual))

# Función principal
def main():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción (1-5): ")

        if opcion == '1':
            comprar_departamento()
        elif opcion == '2':
            mostrar_departamentos_disponibles()
        elif opcion == '3':
            ver_listado_compradores()
        elif opcion == '4':
            mostrar_ganancias_totales()
        elif opcion == '5':
            salir()
            break
        else:
            print("Opción no válida. Intente nuevamente.")

if __name__ == "__main__":
    main()
```

```python
import datetime


departamentos_disponibles = [['A'] * 4 for _ in range(10)]
compradores = []
precios = {'A': 3800, 'B': 3000, 'C': 2800, 'D': 3500}

def mostrar_menu():
    print("Menú:")
    print("1. Comprar departamento")
    print("2. Mostrar departamentos disponibles")
    print("3. Ver listado de compradores")
    print("4. Mostrar ganancias totales")
    print("5. Salir")

def comprar_departamento():
    try:
        piso = int(input("Ingrese el piso (1-10): "))
        tipo = input("Ingrese el tipo de departamento (A/B/C/D): ").upper()

        if piso < 1 or piso > 10 or tipo not in ['A', 'B', 'C', 'D']:
            print("Entrada inválida. Verifique piso y tipo de departamento.")
            return

        departamento = f"{tipo}{piso}"
        if departamentos_disponibles[piso - 1][ord(tipo) - ord('A')] == 'X':
            print("Departamento no disponible.")
            return

        run_comprador = input("Ingrese el RUN del comprador (sin guiones ni puntos): ")
        if not run_comprador.isdigit():
            print("RUN debe contener solo dígitos.")
            return

        compradores.append({'run': run_comprador, 'departamento': departamento})
        departamentos_disponibles[piso - 1][ord(tipo) - ord('A')] = 'X'
        print("Operación realizada correctamente.")

    except ValueError:
        print("Entrada inválida. Piso debe ser un número.")

def mostrar_departamentos_disponibles():
    print("Estado actual de la venta de departamentos:")
    for i in range(10):
        for j in range(4):
            estado = 'X' if departamentos_disponibles[i][j] == 'X' else '-'
            print(f"Tipo {chr(ord('A') + j)}{i + 1}: {estado}", end="   ")
        print()

def ver_listado_compradores():
    print("Listado de compradores:")
    for comprador in sorted(compradores, key=lambda x: x['run']):
        print(f"RUN: {comprador['run']}, Departamento: {comprador['departamento']}")

def mostrar_ganancias_totales():
    print("Ganancias totales:")
    total_por_tipo = {'A': 0, 'B': 0, 'C': 0, 'D': 0}

    for comprador in compradores:
        tipo = comprador['departamento'][0]
        total_por_tipo[tipo] += precios[tipo]

    for tipo, total in total_por_tipo.items():
        cantidad = departamentos_disponibles[0].count(tipo)
        print(f"Tipo {tipo}: {precios[tipo]} UF {cantidad} {total} UF")

    total_general = sum(total_por_tipo.values())
    print(f"TOTAL {len(compradores)} {total_general} UF")

def salir():
    print("Saliendo del sistema. Gracias por usar Casa Feliz.")
    print("Fecha:", datetime.date.today())
    exit()

def main():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción (1-5): ")

        if opcion == '1':
            comprar_departamento()
        elif opcion == '2':
            mostrar_departamentos_disponibles()
        elif opcion == '3':
            ver_listado_compradores()
        elif opcion == '4':
            mostrar_ganancias_totales()
        elif opcion == '5':
            salir()
        else:
            print("Opción no válida. Por favor, elija una opción del 1 al 5.")

if __name__ == "__main__":
    main()
```
